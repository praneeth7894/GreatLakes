# E. Distribution for BMI, age and charges

```python
bmisorted= sorted(bmi)
agesorted= sorted(age)
chargessorted= sorted(charges)
```

```python
#Normal Distribution
bmifit= stats.norm.pdf(bmisorted, np.mean(bmisorted), np.std(bmisorted))
plt.plot(bmisorted,bmifit,'-o')
plt.hist(bmisorted,normed=True)
plt.show()
```

```python
agefit= stats.norm.pdf(agesorted, np.mean(agesorted), np.std(agesorted)) 
plt.plot(agesorted,agefit,'-o') 
plt.hist(agesorted,normed=True) 
plt.show()
```

```python
chargesfit= stats.norm.pdf(chargessorted, np.mean(chargessorted), np.std(chargessorted)) 
plt.plot(chargessorted,chargesfit,'-o') 
plt.hist(chargessorted,normed=True) 
plt.show()
```
